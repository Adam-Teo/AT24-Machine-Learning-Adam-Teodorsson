````python
# 0) a
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

#help(keras.datasets.mnist.load_data)

(Xtr, ytr), (Xte, yte) = keras.datasets.mnist.load_data()

# assert throws an error if the condition is not true
# First we have the number of rows aka observation in 
# this case images, in X_train there are 60k of images
# then we have the second and third tensor with 28  
# dimension each
assert Xtr.shape == (60000, 28, 28)
assert Xte.shape == (10000, 28, 28)
assert ytr.shape == (60000,)
assert yte.shape == (10000,)

def pick_sample(Xtr, ytr, size):
    np.random.seed(42)
    sample_idx = list()
    while len(sample_idx) < size:
        sample_idx = np.random.choice(np.arange(0,Xtr.shape[0]), size=size, replace=False)
    return np.array([Xtr[idx] for idx in sample_idx]), np.array([ytr[idx] for idx in sample_idx])

Xsam, ysam = pick_sample(Xtr, ytr, 20)
print( Xsam.shape, ysam.shape )

plt.imshow(Xsam[0], cmap=plt.cm.gray_r, interpolation="nearest")

# Reshaping the (60000, 28, 28)  to (60000, 784)
Xtr = Xtr.reshape(Xtr.shape[0], -1)
Xte = Xte.reshape(Xte.shape[0], -1)
````

````python
fig, axes = plt.subplots(nrows=1, ncols=10, figsize=(10,3))
for ax, image, label in zip(axes, Xsam, ysam):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
    ax.set_title(f"Label {label}")

fig, axes = plt.subplots(nrows=1, ncols=10, figsize=(10,3))
for ax, image, label in zip(axes, Xsam[9:], ysam[9:]):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
    ax.set_title(f"Label {label}")

````

## 1. Train|test|validation split (*)

As you already have training data and testing data, do a split on the training data to get validation data. The validation data will be used for tuning $k$ number of neighbors for KNN. I used 0.16 for validation as it is close to 1/6 to get around 10000 validation samples.

---

````python
from sklearn.model_selection import train_test_split 
Xtr_ttv, Xva_ttv, ytr_ttv, yva_ttv = train_test_split( Xtr, ytr, test_size=0.16, random_state=42 )

Xtr_ttv.shape, Xva_ttv.shape, ytr_ttv.shape, yva_ttv.shape
````

## 2. Hyperparameter tuning (*)

Train the models for a set of $k$-values using the training data and make predictions on validation data. Plot the validation accuracy against different $k$-values. Based on your plot, which $k$ do you choose? 

<details>

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

error_list = list()


for k in range(1,5):
    model_KNN = KNeighborsClassifier(n_neighbors=k)
    model_KNN.fit(Xtr_ttv, ytr_ttv)
    y_hat = model_KNN.predict(Xva_ttv)
    error_list.append(1-accuracy_score(yva_ttv, y_hat))

fig, ax = plt.figure(), plt.axes()
ax.plot(range(1, len(error_list)+1), error_list, ".-")
ax.set(title="Elbow plot", xlabel = "k neightbours", ylabel= "Error")
````
